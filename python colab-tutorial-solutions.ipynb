##Dictionary

````python
# Calculate the total price of items in a shopping cart
cart = {
    "apple": 1.0,
    "banana": 0.5,
    "orange": 1.2,
    "grapes": 2.5
}
quantities = {
    "apple": 2,
    "banana": 3,
    "orange": 1,
    "grapes": 0.5
}

total_price = 0

# Use a loop to calculate the total price of items in the cart, considering their quantities
# YOUR CODE HERE

print(total_price)
````

Solutions:
```
for item, price in cart.items():
    total_price += price * quantities.get(item, 0)
```
